# patterns in python
patterns are created using for loops, multiple for loops are used to print patterns


1.   outer loop to print number of rows
2.   inner loop to print number of columns 
3.   a variable to print whitespace according to the required place

```python
n=int(input("Enter no of rows "))
# outer loop to go through rows
for i in range(0,n):
  # inner loop to handle number of columns
  for j in range(0,i+1):
    # printing star
    print("* ",end="")
    # ending line after each row
  print()
```

```python
n=int(input("Enter num: "))

def palindrome(n):
  num_str=str(n)
  rev_str=num_str[::-1]
  if num_str==rev_str:
    return True
for i in range(n,1,-1):
  if palindrome(i):
    print("Largest palindrome is ",i)
    break
```
